```python
import os

from langchain_community.document_loaders import PyPDFLoader

from langchain_community.vectorstores import Chroma

from langchain_openai import OpenAI
```

```python
loader = PyPDFLoader("/content/Sai Tapovanam.pdf") #upload the document you wish to use for chatbot
pages = loader.load()
```

```python
len(pages)
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter

def split_docs(document):
  text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=100)
  docs = text_splitter.split_documents(document)
  return docs
```

```python
new_pages = split_docs(pages)
len(new_pages)
```

```python
new_pages[500].page_content
```

```python
from langchain_community.embeddings.sentence_transformer import (
    SentenceTransformerEmbeddings,
)

embedding_function = SentenceTransformerEmbeddings(model_name="all-MiniLM-L6-v2")

db = Chroma.from_documents(new_pages, embedding_function)
```

```python
from langchain_together import Together


llm = Together(
    model="meta-llama/Llama-2-70b-chat-hf",
    max_tokens=256,
    temperature=0,
    top_k=1,
    together_api_key="6b0599d6baf7f833ddeea4165e07a17a4fccf1298e4a87d1581cc56afa00bf4a"
)
```

```python
retriever = db.as_retriever(similarity_score_threshold = 0.9)
```

```python
from langchain.prompts import PromptTemplate
prompt_template = """Please answer questions related to Sathya Sai Baba and a book called Tapovanam. Try explaining in simple words. Answer in less than 100 words. If you don't know the answer simply respond as "I am not aware!"
 CONTEXT: {context}
 QUESTION: {question}"""

PROMPT = PromptTemplate(template = f"[INST] {prompt_template} [/INST]", input_variables=["context", "question"])
```

```python
from langchain.chains import RetrievalQA
chain = RetrievalQA.from_chain_type(
    llm = llm,
    chain_type='stuff',
    retriever= retriever,
    input_key = 'query',
    return_source_documents = True,
    chain_type_kwargs={"prompt":PROMPT},
    verbose=True

)
```

```python
query = "who is sathya sai baba"
response = chain(query)
response['result']
```
